```python
import pandas as pd

# ## 01. Data Exploration

from src.data_loader import load_titanic_data

# Load data
train_df, test_df = load_titanic_data("V:/titanic")

# Quick peek
display(train_df.head())
print(train_df.info())
```

```python
train_df.head()
```

```python
# Summary for numerical features
display(train_df.describe())

# Info & missing-value counts
print(train_df.info())
print("\nMissing values per column:")
print(train_df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Example: Age distribution
plt.figure()
train_df['Age'].hist(bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Example: Count of Survivors vs. Non‑survivors
plt.figure()
train_df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Counts (0 = No, 1 = Yes)")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
# Pclass vs. Survival rate
pclass_survival = train_df.groupby('Pclass')['Survived'].mean()
display(pclass_survival)

# Sex vs. Survival rate
sex_survival = train_df.groupby('Sex')['Survived'].mean()
display(sex_survival)
```
